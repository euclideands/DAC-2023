## **Handling Missing Values**

```python
# Initialize a dictionary to store missing value representations
missing_value_representations = ['', ' ', '-', 'NA', 'N/A', '999', '-999', None, np.nan, '*', '9999', '99999']

# Initialize a set to store unique types of missing values
unique_missing_value_types = set()

# Iterate through columns and check for types of missing values
for column in df.columns:
    for value in df[column]:
        if pd.isna(value):
            value_str = 'NaN'
        else:
            value_str = str(value)
        
        if value_str in missing_value_representations:
            unique_missing_value_types.add(value_str)

# Display unique types of missing values found in the entire DataFrame
if not unique_missing_value_types:
    print("No missing values found in the DataFrame.")
else:
    print(f"Jenis-jenis missing value dalam DataFrame: {', '.join(unique_missing_value_types)}")

```

Smurf attack is the popular type of attack sending ICMP echo request packet to middle device. ICMP packets have source address name as victimâ€™s IP address and middle device address as destination address. DoS attack reduces the performance of server by overflowing the ICMP traffic as discussed above in the Neptune and Smurf attack.
